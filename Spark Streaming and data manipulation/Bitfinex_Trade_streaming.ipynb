```python
import numpy as np
import pandas as pd

data_array = np.load('bitfinex_trade_array.npy',allow_pickle=True)

print("### 10 million points of data ###")
print("\nData summary:\n", data_array)
print("\nData shape:\n", data_array.shape)

# id, MTS millisecond time stamp, +- Amount positive bought, negative sold, Price
pd_df = pd.DataFrame({
    "ID": data_array[:, 0], 
    'MTS': pd.Series(data_array[:, 1],dtype = int), 
    "AMOUNT" : data_array[:,2], 
    "PRICE": data_array[:,3]
})

pd_df.info()

```

```python
pd_df.ID = pd_df.ID.astype('int')

data_array[0][1]
```

```python
import numpy as np
import pandas as pd

order_array = np.load('binance_order_array.npy',allow_pickle=True)

print("\nData summary:\n", order_array)
print("\nData shape:\n", order_array.shape)
```

```python
from pyspark import SparkContext
from pyspark.streaming import StreamingContext

try:
    sc = SparkContext('local', 'Pyspark demo')
    ssc = StreamingContext(sc, 1)
except ValueError:
    print('SparkContext already exists!')
from pyspark.sql import SparkSession
try:
    spark = SparkSession.builder.appName('Demo').getOrCreate()
except ValueError:
    print('SparkSession already exists!')
```

```python
spark_df = spark.createDataFrame(pd_df)
spark_df
```

```python
bit_finex = ssc.sparkContext.parallelize (
    data_array
)

order = ssc.sparkContext.parallelize (
    order_array
)
```

```python
finex_map = bit_finex.map(lambda x: (str(x[1])[:-5], [x[1],x[3]]))
```

```python
finex_map.take(100)
```

```python
def seqFunc(acc, row):
    if(acc[0] > row[0]):
        return acc 
    else: 
        return row

def combFunc(acc1, acc2):
    if(acc1[0] > acc2[0]):
        return acc1
    else:
        return acc2

finex_Max = finex_map.aggregateByKey((0,0), seqFunc, combFunc).map(lambda x: (x[0], x[1][1]))

finex_Max.take(10)
```

```python
order_map = order.map(lambda x: (str(x['time'])[:-3],0))
```

```python
order_map.take(5)
```

```python
joined = order_map.leftOuterJoin(finex_Max)
```

```python
joined.take(20)
```

```python
dropped = joined.map(lambda x: [x[0],x[1][1]])
```

```python
dropped.take(20)
```

```python
def func(iterator):
    last = None
    
    out = []
    
    for x in iterator:
        if x[1] is None:
            x[1] = last
        else:
            last = x[1]
        out.append([x[0],x[1]])
        
    for x in reversed(out):
        if x[1] is None:
            x[1] = last
        else:
            last = x[1]
    return out 

filledNone = dropped.mapPartitions(func)

filledNone.take(100)
```

```python
def arranger(iterator):
    sv = 0 
    out = []
    res = []

    for i in range(np.array(iterator).shape[0] - 1):
        iterator[i][1] = iterator[i+1][1]
        out.append([iterator[i][0],iterator[i][1]])
    return out

arranged = filledNone.mapPartitions(arranger)

print(filledNone.count())
arranged.count()
```

```python
from influxdb import InfluxDBClient
```

```python
client = InfluxDBClient(host='localhost', port=8086)
```

```python
client.create_database('bitfinex_trade')
```

```python
client.get_list_database()
```

```python
def influx(partitions):
    result = []
    for i in range(np.array(partitions).shape[0] - 1):
        result.append('bitfinex_trade timestamp='+partitions[i][0]+','+'price='+ str(partitions[i][1]))
    return result

inf_final=arranged.mapPartitions(influx)  
```

```python
inf_final.take(5)
```

```python
client.write_points(inf_final.collect(), database='bitfinex_trade', time_precision='ms', batch_size=1, protocol='line')
```

```python
client.switch_database('bitfinex_trade')
```
